# Часто используемые чаще всего приемы языка:

1. Разделителем абзацев служит пустая строка;

2. Можно использовать HTML теги, например (<strong>текст</strong> или **текст**);

3. *ШРИФТ*: Наклонное и полужирное начертание в Markdown задаются при помощи символов * и _:

    3.1 один символ — для наклонного текста (_italic_, *italic*);

    3.2 два символа — для жирного текста (__strong__, **strong**);
    
    3.3 три — для наклонного и жирного одновременно (***жирный и наклонный***).

4. *ЗАГОЛОВКИ* - отличаются лишь # от 1 до 6:

    # Заголовок H1
    ## Заголовок H2
    ### Заголовок H3
    #### Заголовок H4
    ##### Заголовок H5
    ###### Заголовок H6

5. Для центрирования заголовка можно воспользоваться тегом:  <center>**Мой заголовок**</center>

6. *СПИСКИ*: Для разметки ненумерованных списков можно использовать символы *, - или + — результат будет один:
    + список 2;
        + подсписок 2.1
    - список;
    * список.

7. *ССЫЛКИ*: 
    - без подсказки — [текст ссылки](http://example.com/link);
    - c подсказкой — [текст ссылки](http://example.com/link "Подсказка")

    - перед отображением изображения ставится (воскл.знак) ![тут будет помещено изображение](https://i.imgur.com/3uj9teq.png)

    другой вариант использовать HTML:

    <img src=https://i.imgur.com/3uj9teq.png width=500px height=30%>

8. *КОД И ЦИТИРОВАНИЕ*: Для выделения программного кода используется обратный апостроф:

    - одинарный парный — для вставки строки кода в текст;

    `print('Hello World!')`

    - двойной парный — для вставки небольшого участка кода, содержащего одинарный апостроф, в текст;

    ``  num = 13

        multiply = num + 55

        print(f'The result is: {multiply}')
    ``
    
    - тройной парный — для вставки блока программного кода (еще можно указать язык программирования чтобы автоматически подсветить участки кода, как в редакторе.)


# ФОРМУЛЫ

Когда в проекте есть математическая составляющая, важно отразить её в описании.

Вставка математических формул в Markdown осуществляется с помощью специальной библиотеки KaTeX. 

Чтобы начать использовать KaTeX в Markdown, необходимо воспользоваться символом $. Если обрамить формулу с обеих сторон одним символом $, то её можно встроить в текст, а если двумя — формула автоматически центрируется.

Пусть задано выражение:

$$ a = b + c, $$

где $ a = 0 $

Латинские буквы набираются с клавиатуры, а для греческих есть специальные команды, которые выражаются через символ \

* $\alpha$ - `$\alpha$`

* $\omega$ - `$\omega$`

* $\sigma$ - `$\sigma$`

* $\delta$ - `$\delta$`

* $\epsilon$ - `$\epsilon$`

Степени и индексы набираются с помощью символов ^ и _ соответственно. Если символов, которые нужно поместить в степень или индекс, несколько, то они выделяются фигурными скобками. 

Например:

* $a^2$ - `$a^2$`

* $b_{ij}$ - `$b_{ij}$`

* $w^{ij}_n$ - `$w^{ij}_n$`

Для того чтобы создать «двухэтажную» дробь, можно воспользоваться оператором \frac с двумя параметрами, которые передаются в фигурных скобках (числитель и знаменатель). 

Например:

$\frac{1+x}{n}$ - `$\frac{1+x}{n}$`
