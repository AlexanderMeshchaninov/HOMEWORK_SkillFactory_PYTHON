## DENSE_RANK()

Пример определение лучших сотрудников по департаментам с присвоением ранга (rnk)

C использование конструкции dense_rank() over (partition by … order by …) rnk
Допустим, у нас есть таблица employees с данными о производительности сотрудников:

employee_id	department	employee_name	performance_score
1	HR	John	85
2	HR	Jane	90
3	HR	Jack	85
4	IT	Alice	95
5	IT	Bob	90
6	IT	Charlie	85
Задача: присвоить плотные ранги сотрудникам по их производительности в каждом департаменте.

``` sql
select    
    department,
    employee_name,
    performance_score,
    dense_rank() over (partition by department order by performance_score desc) rnk
from
    employees
```

Результат:

department	employee_name	performance_score	rnk
HR	Jane	90	1
HR	John	85	2
HR	Jack	85	2
IT	Alice	95	1
IT	Bob	90	2
IT	Charlie	85	3

Объяснение:

Сотрудники разделены по департаментам.
В каждом департаменте сотрудники упорядочены по производительности в порядке убывания.
Сотрудники с одинаковыми производственными оценками получают одинаковый ранг, следующий уникальный ранг не пропускается (плотный ранг).
Таким образом, конструкция dense_rank() over (partition by ... order by ...) позволяет присваивать плотные ранги строкам в пределах разделов, упорядочивая их по заданному критерию. Эта функция полезна для определения позиций или рангов объектов в разных группах, таких как классы, месяцы или департаменты.

## Ранжирование продаж продуктов по месяцам

Допустим, у нас есть таблица sales с данными о продажах:

sale_id	month	product_name	amount
1	Jan	Product A	100
2	Jan	Product B	200
3	Jan	Product C	150
4	Feb	Product A	120
5	Feb	Product B	80
6	Feb	Product C	160
Задача: присвоить плотные ранги продуктам по объему продаж в каждом месяце.

``` sql

select
    month,
    product_name,
    amount,
    blank_rank() over (partition by month order by amount desc) rnk
from
    sales
    
```
Результат:

month	product_name	amount	rnk
Jan	Product B	200	1
Jan	Product C	150	2
Jan	Product A	100	3
Feb	Product C	160	1
Feb	Product A	120	2
Feb	Product B	80	3

Объяснение:

Продукты разделены по месяцам.
В каждом месяце продукты упорядочены по объему продаж в порядке убывания.
Каждый продукт получает плотный ранг в пределах месяца.
